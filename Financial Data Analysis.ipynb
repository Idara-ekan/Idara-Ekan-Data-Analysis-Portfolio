```python
index_list = list(clean_data.index)
index_list
```

```python
budget = 0
for n in index_list:
    if n in main:
        pass
    else:
        budget += data['Fund allotted(in ₹crores)'].iloc[n]
print(budget)
```

```python
data1 = clean_data.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': budget}
data1 = data1.append(row, ignore_index = True)
print(data1)
```

```python
data1.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
data1.plot.bar(x = 'Department /Ministry', y = 'Fund allotted(in ₹crores)')
```

```python
df = data1["Fund allotted(in ₹crores)"]
labels = data1["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0,0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of the Budget", fontsize=20)
plt.show()
```
